## Introduction and Background

In this laboratory you will explore the sources of uncertainties in the measurements using digital multi-meters or DMM.

In DMM, the accuracy is given in the datasheet, Appendix A, and the resolution is expressed in significant digit. For the Agilent 34405A used in this lab, the resolution is 5.5 digits. The first “5” represents full digits, i.e. they can display from 0-9. The “0.5” represents the first digit of the reading. In this case, 0.5, means the first digit can represent 0 or 1. Therefore, the number of counts for this multimeter is 200,000 and its resolution is ± ½ the last digit.

Evaluating sources of uncertainties, or doubts, in datasets is very important to have established level of confidence in the reported data. The most obvious and significant sources of doubt are: inaccuracy of the calibrator’s output value, repeatability in the measurements, and resolution of the instrument. However, many other factors contribute also to the overall uncertainty: ambient temperature effects, electromagnetic noise, operator errors, etc.

For evaluating uncertainties in measurements, rules have been established internationally. You will follow them in this lab. The detailed rules can be found in the “Guide to the expression of Uncertainty in Measurement”, or “GUM” [1]. A shortened version can be in found in [2].

**Definition**

Following the GUM nomenclature, two types of uncertainties are defined:<br><p></p>
- Type A: They are evaluated by statistical analysis of series of observations<br><p></p>
- Type B: They are evaluated by means others than the statistical analysis of series of observations.<br><p></p>
- If uncertainties are not Type A, they are Type B.<br><p></p>

For example, Type B uncertainties are evaluated by analytical model of a transducer and its boundary conditions or from calibration certificates. Type A and B differ from random and systematic uncertainties traditionally used in engineering.

**Standard uncertainty, $u_i$**: uncertainty of the result of a measurement expressed as a standard deviation. It can be thought as ± one standard deviation.

$$u_i$$

**Combined standard uncertainty, $u_m$**: standard uncertainty of the result of a measurement when that result is obtained from the values of a number of other quantities. It is calculated as root sum of the squares of the elemental standard uncertainties and pertains to ± one standard deviation (63% coverage factor):

$$u_c = \sqrt{u_1^2 + u_2^2 + ... + u_N^2}$$

One should be careful when using this formula that all the standard uncertainties have the same units and coverage factor, or one will introduce errors.

**Expanded uncertainty, $u_m$**: Usually, uncertainty is reported with a larger confidence level, such as 95% or 99%. The expanded uncertainty is related to the combined standard uncertainty by a **coverage coefficient, $k$**:

$$U_m = k u_c$$

If you have a large sample, e.g. more than 30 independent measurements, the coverage coefficient is given by the normal distribution. In this case a $k = 2$ for 95% confidence level. However, if you have 30 or less measurements, your coverage factor will be given by the Student t-distribution, which was introduced in the statistics review lecture. You will also need materials covered in that lecture for the remainder of the lab.

In this lab you will see how to calculate the two types of uncertainties, how to allocate a value for the coverage coefficient, and how to improve the accuracy of an instrument by calibrating it.

### Part II - Type B Uncertainties

Type B uncertainties encompass all other uncertainties that cannot be determined statistically during the measurement process. They include, but are not limited to: calibration inaccuracy, limited instrument resolution, environment effects, etc. Here you will explore in details two types of doubts in your measurements: range and reading uncertainties and limitation in the DMM resolution.

In accordance with the convention adopted above, $u_i$ is a standard uncertainty and is defined at ± one standard deviation (i.e. 68% confidence level). In some manufacturer, the reported accuracies can be at 90%, 95%, 99%, etc, which implies that you will have to adjust the reported value to be able to work on your uncertainties at ± one standard deviation.

If you do not have sufficient information to know for sure, you should treat the pdf as a rectangular or uniform distribution, ie values are equally likely. In cases you are only able to estimate the upper and lower limits of uncertainty, the standard uncertainty is:

$$u_i = \frac{a}{\sqrt{3}}$$

With $a$ the half-width between the lower and upper limits. Here you are lucky, Agilent reports the uncertainties of its instruments to 68% confidence level.
